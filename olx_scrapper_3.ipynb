```python
import requests
import json
import pandas as pd
```

```python
listkecamatan = [
                 '5001299',
                 '5001296',
                 '5001302',
                  '5001297',
                  '5001301',
                  '5001304',
                  '5001298',
                  '5001300',
                  '5001303',
                  '5001306',
                  '5001295',
                  '5001305'
]
```

```python
olx_url = 'https://www.olx.co.id/'
keyword = 'rumah'

header = {
        'user-agent': 'Chrome', 
        'referer' : '{}search?keyword={}'.format(olx_url, keyword)
    }

judul, loc_kecamatan, harga, l_tanah, l_bangunan, kamar_tidur, kamar_mandi= [], [], [], [], [], [], []

for kecamatan in listkecamatan:
  for page in range(0,50):

    url = 'https://www.olx.co.id/api/relevance/v2/search?category=5158&facet_limit=100&location={}&location_facet_limit=20&platform=web-desktop&query=rumah&spellcheck=true&type=rumah&page={}'.format(kecamatan,page)

    #API Req
    r = requests.get(url, headers=header).json()

    #Scrapping

    for item in r['data']:
      judul.append(item['title'])
      loc_kecamatan.append(item['locations_resolved']['SUBLOCALITY_LEVEL_1_name'])
      l_tanah.append(item['parameters'][2]['formatted_value'])
      harga.append(item['price']['value']['display'])
      l_bangunan.append(item['parameters'][1]['formatted_value'])
      kamar_tidur.append(item['parameters'][3]['formatted_value'])
      kamar_mandi.append(item['parameters'][4]['formatted_value'])
```

```python
len(judul)
```

```python
#Save the data
listcols = ['judul', 'kecamatan', 'luas tanah (m2)', 'luas bangunan (m2)', 'kamar tidur', 'kamar mandi', 'price' ]
dictdata = dict(zip(
    listcols,
    (judul,
     loc_kecamatan,
     l_tanah,
     l_bangunan,
     kamar_tidur,
     kamar_mandi,
     harga)
))
```

```python
with open('olxhouse.json', 'w') as fp:
  json.dump(dictdata, fp)
```

```python
df = pd.DataFrame(data=dictdata)
df
```
